```python
#remove single character mostly punctuation
import string
import nltk
from nltk.stem import PorterStemmer

clean_data1 = reference_text.response.str.cat(sep=' ')

words0 = nltk.word_tokenize(clean_data1)


words1 = [word for word in words0 if len(word) > 2]


#Remove numbers
words2 = [word for word in words1 if not word.isnumeric()]

#Remove non breaking space
words3 = [word for word in words2 if word != "nbsp"]

# Stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
words4 = [w for w in words3 if not w in stop_words]

#lowercase all words
words5 = [word.lower() for word in words4]

exclude = set(string.punctuation)
words6 = ' '.join(ch for ch in words5 if ch not in exclude)


for char in string.punctuation:
    words7 = words6.replace(char, ' ')

words7 = words7.replace('.', ' ')

```

```python
import nltk
from nltk.stem import WordNetLemmatizer

words8 = nltk.word_tokenize(words7)
lemmatizer = WordNetLemmatizer()
words9 = []
for word in words8:
    wor=lemmatizer.lemmatize(word)
    words9.append(wor)
    print(lemmatizer.lemmatize(word))

   
```

```python
words9
```

```python
#Calculating FWR
from nltk.probability import FreqDist

FWR = FreqDist()

for words in words9:
    FWR[words.lower()]+=1
FWR
```

```python
   
df_fdist = pd.DataFrame.from_dict(FWR, orient='index')
df_fdist.columns = ['Fwr']
df_fdist.index.name = 'Word'
df_fdist


    
```

```python
# compute p(r|w) = f_wr / sum(f_wr)_{for all r}

df_fdist['Pwr'] = df_fdist['Fwr'] /  df_fdist['Fwr'].sum()

df_fdist
```
