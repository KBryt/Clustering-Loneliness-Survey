```python
clean_data = raw_data[raw_data['response'] != '-99']
```

```python
clean_data['response'] = clean_data['response'].str.lower()
clean_data['response'] = clean_data['response'].str.replace('loneliness', '')
clean_data['response'] = clean_data['response'].str.replace('loneiness', '')
clean_data['response'] = clean_data['response'].str.replace('loneness', '')
clean_data['response'] = clean_data['response'].str.replace('aloneliness', '')
clean_data['response'] = clean_data['response'].str.replace('loneness', '')
clean_data['response'] = clean_data['response'].str.replace('/', ' ')
clean_data['response'] = clean_data['response'].str.replace('\n', ' ')
clean_data['response'] = clean_data['response'].str.replace('.', '')
clean_data['response'] = clean_data['response'].str.replace(',', '')
                                                            
```

```python
def getAnalysis(score):
    if score <= 19:
        return '15-19'
    elif score <=39:
        return '20-39'
    elif score <=49:
        return '40-49'
    elif score <=59:
        return '50-59'
    elif score <=69:
        return '60-69'
    else:
        return '70 and above'
    
def getAnalysis1(score):
    if score == 1:
        return 'Male'
    elif score ==2:
        return 'Female'
    elif score ==3:
        return 'Others'
    else:
        return 'Prefer not to say'
    
clean_data['age_group'] = clean_data['age'].apply(getAnalysis)

    
clean_data['gender_cat'] = clean_data['gender'].apply(getAnalysis1)
#clean_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import re


import pandas 

corpus = words10
# View most frequently occuring keywords
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in      
                   vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], 
                       reverse=True)
    return words_freq[:n]

# Convert most freq words to dataframe for plotting bar plot, save as CSV
top_words = get_top_n_words(corpus, n=50)
top_df = pandas.DataFrame(top_words)
top_df.columns=["Keyword", "Frequency"]
print(top_df)
#top_df.to_csv(file_prefix + '_top_words.csv')

# Barplot of most freq words
import seaborn as sns
sns.set(rc={'figure.figsize':(13,8)})
g = sns.barplot(x="Keyword", y="Frequency", data=top_df, palette="rainbow")
g.set_xticklabels(g.get_xticklabels(), rotation=45)
```
