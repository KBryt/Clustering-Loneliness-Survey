# Creating a 3D plot in Matplotlib

```python
# Import libraries
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
# Creating our figure and axes
fig = plt.figure(figsize=(4,4))

ax = fig.add_subplot(111, projection="3d") # ax will be used to add any plot to the function
```

```python
# Plot the point
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(2,3,4) # plot the point (2, 3, 4) on the figure
plt.show
```

```python
# Customizing a 3D plot
import numpy as np
x = np.linspace(-4*np.pi, 4*np.pi, 50)
y = np.linspace(-4*np.pi, 4*np.pi, 50)
z = x**2 + y**2
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(x, y, z)
ax.set_title("Distribution") # Additing a titleax

# setting the label for x, y, z
ax.set_xlabel("Age")
ax.set_ylabel("Gender")
ax.set_zlabel("Education")

# Changing marker and color
ax.scatter(x, y, z, marker='*', c="red")


plt.show()
```

```python
# we modify our model 
np.random.seed(42)
ages = np.random.randint(low=8, high=30, size=35)
heights = np.random.randint(130, 195, 35)
weights = np.random.randint(30, 160,  35)
fig = plt.figure(figsize=(6,6)) # you can change the size of the figure by `odifying the figsize`



ax = fig.add_subplot(111, projection="3d")
ax.scatter(xs=heights, ys=weights, zs=ages)
ax.set_title("Age-wise body weight-height distribution")
ax.set_xlabel("Height (cm)")

# Modifying the min_limit and max_limit
ax.set_xlim(100, 200)
ax.set_ylim(20, 160)
ax.set_zlim(5, 35)

#modifying the ticks for the axis
ax.set_xticks([100,125,150,175,200])
ax.set_yticks([20,55,90,125,160])
ax.set_zticks([5,15,25,35])

# Naming axes
ax.set_ylabel("Weight (kg)")
ax.set_zlabel("Age (years)")

# Turning on/of the grid. Set it to false if you need the grid
ax.grid(False)


plt.show()
```

```python
np.random.seed(42)
ages = np.random.randint(low = 8, high = 30, size=35)
heights = np.random.randint(130, 195, 35)
weights = np.random.randint(30, 160, 35)

gender_labels = np.random.choice([0, 1], 35) # we modify the color of gender by 0 for male, 1 for female

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs = heights, ys = weights, zs = ages, c=gender_labels)

ax.set_title("Age-wise body weight-height distribution")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")

ax.set_zlabel("Age (years)")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scat_plot = ax.scatter(xs=heights, ys=weights, zs=ages, c=gender_labels)
color_bar = plt.colorbar(scat_plot, pad=0.2)
color_bar.set_ticks([0, 1])

color_bar.set_ticklabels(["Male", "Females"])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#plot scatter plots for the 3 states and add a legend to distinguish them from each other.
# create the 3 plots in a for-loop and assign a different label to them each time
labels = ["Florida", "Georgia", "California"]

for l in labels:

    ages = np.random.randint(low = 8, high = 20, size=20)

    heights = np.random.randint(130, 195, 20)

    weights = np.random.randint(30, 160, 20)

    ax.scatter(xs = heights, ys = weights, zs = ages, label=l)



ax.set_title("Age-wise body weight-height distribution")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")


ax.set_zlabel("Age (years)")
ax.legend(loc="best")
plt.show()
```

```python
# Modifying the points by altering the size of the data
np.random.seed(42)

ages = np.random.randint(low = 8, high = 30, size=35)
heights = np.random.randint(130, 195, 35)
weights = np.random.randint(30, 160, 35)

bmi = weights/((heights*0.01)**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs = heights, ys = weights, zs = ages, s=bmi*5 )

ax.set_title("Age-wise body weight-height distribution")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")

ax.set_zlabel("Age (years)")
plt.show()
```

```python
# Here we plot the normal distribution function in 3D
from scipy.stats import multivariate_normal

X = np.linspace(-5,5,50)
Y = np.linspace(-5,5,50)

X, Y = np.meshgrid(X,Y)
X_mean = 0; Y_mean = 0

X_var = 5; Y_var = 8
pos = np.empty(X.shape+(2,))

pos[:,:,0]=X
pos[:,:,1]=Y

rv = multivariate_normal([X_mean, Y_mean],[[X_var, 0], [0, Y_var]])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, rv.pdf(pos), cmap="plasma")
plt.show()
```

```python
# Polygon in 3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 0, 3, 4]
y = [0, 5, 5, 1]
z = [1, 3, 4, 0]

vertices = [list(zip(x,y,z))]
poly = Poly3DCollection(vertices, alpha=0.8)

ax.add_collection3d(poly)
ax.set_xlim(0,5)
ax.set_ylim(0,5)
ax.set_zlim(0,5)    
```
